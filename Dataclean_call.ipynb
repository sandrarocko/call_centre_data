### Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel(r'/Users/sandra/Documents/Sql/Customer Call List.xlsx')
```

##### Drop duplicates

```python
df1 = df.drop_duplicates()
```

```python
df1 = df1.drop(columns='Not_Useful_Column')
```

##### We have values, '/'  White, _ and Flanderson, '...' and with Potter

```python
df1['Last_Name'] = df1['Last_Name'].str.lstrip('/')
```

```python
df1
```

```python
df1['Last_Name'] = df1['Last_Name'].str.lstrip('...')
```

```python
df1
```

##### Cleaning "Phone number" column for it to look exatly the same

```python
df1["Phone_Number"] = df1["Phone_Number"].astype(str)
```

```python
df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
```

```python
df1
```

```python
#df1['Phone_Number'].apply(lambda x: x[0:3]+'-'+[3:6]+'-'[6:10])
#str(df1['Phone_Number'])
df1['Phone_Number'] = df1['Phone_Number'].apply(lambda x: str(x))
df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].apply(lambda x: x[0:3]+'-'+ x[3:6] + '-' + x[6:10])
df1
```

```python
df1['Phone_Number'] = df1['Phone_Number'].str.replace('nan--','')
```

```python
df1['Phone_Number'] = df1['Phone_Number'].str.replace('Na--','')
```

```python
df1
```

##### OK now "Address"

```python
df1[['Street_Address', 'State', 'ZIP Code']] = df1['Address'].str.split(',', 2, expand=True)    #splitting on a ',' a potem ile split
```

```python
df1
```

##### Replace "Paying Customer" i "Do_Not_Contact"

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace('Yes','Y')
```

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace('No','N')
```

```python
df1
```

```python
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('Yes','Y')
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('No','N')
```

```python
df1
```

```python
df1 = df1.replace('N/a','')
```

```python
df1
```

```python
df1 = df1.fillna('')
```

```python
df1
```

##### Delete nunmbers "do not contact" 

```python
# zaczynamy od Do_Not_Contact
for x in df1.index:
    if df1.loc[x, 'Do_Not_Contact'] == 'Y': ##if this value is yes then drop
        df1.drop(x, inplace=True)
```

```python
df1
```

```python
for x in df1.index:
    if df1.loc[x, 'Phone_Number'] == '': #if this value is yes then drop
        df1.drop(x, inplace=True)
```

```python
df1
```

```python
df1 = df1.reset_index()
```
